```python
import matplotlib.pyplot as plt
import torch
import numpy as np
%matplotlib inline
def plot_loss(n):
    y = []
    for i in range(0,n):
        enc = torch.load('epoch_{}.pt'.format(i))
        tempy = list(enc)
        y += tempy
    x = range(0,len(y))
    plt.plot(x, y, '.-')
    plt_title = 'BATCH_SIZE = 64; LEARNING_RATE:0.0002'
    plt.title(plt_title)
    plt.xlabel('per 1000 times')
    plt.ylabel('TRAIN LOSS')
    plt.savefig('loss2.jpg')
    plt.show()


if __name__ == "__main__":
    plot_loss(4)
```
